```python
import matplotlib.pyplot as plt
```

```python
x=[2,6]
y=[4,9]
plt.plot(x,y)
plt.show()
```

```python
x=[2,6]
y=[4,9]
x1=[8,3]
y1=[4,7]
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

```python
x=[2,6]
y=[4,9]
x1=[8,3]
y1=[4,7]
plt.xlabel("time")
plt.ylabel("Distance")
plt.plot(x,y)
plt.plot(x1,y1)
plt.show()
```

```python
x=[2,6]
y=[4,9]
x1=[8,3]
y1=[4,7]
plt.xlabel("time")
plt.ylabel("Distance")
plt.plot(x,y,label="park")
plt.plot(x1,y1,label="river")
plt.legend()
plt.grid(color="y")
plt.show()
```

```python
y=[200,60,65,90]
x=["virat","Dhoni","rahul","sachin"]
x1=[80,67,90,100]
y1=["abc","xyz","pqr","v"]
plt.xlabel("Players")
plt.ylabel("Scores")
plt.bar(x,y)
plt.bar(y1,x1)
plt.legend()
plt.grid(color="y")
plt.show()
```

```python
x=[1,5,8,9,3,4,6,10]
y=[4,9,23,6,9,5,6,8]
plt.scatter(x,y,label="mines",marker="x")
plt.legend()
plt.show()
```

```python
x=[1,5,8,9,3,4,6,10]
y=[4,9,23,6,9,5,6,8]
plt.scatter(x,y,label="water",s=150,c="r")
plt.legend()
plt.show()
```
